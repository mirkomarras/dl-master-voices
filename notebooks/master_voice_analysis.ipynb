```python
import pandas as pd
import numpy as np
import pickle
import os
```

```python
def save_obj(obj, name):
    with open(name, 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name):
    with open(name, 'rb') as f:
        return pickle.load(f)
```

```python
base_path = '/beegfs/mm11333/dl-master-voices/data/vs_imps'
```

```python
results = {}
for file in os.listdir(base_path):
    print(file.split('_')[0], file.split('_')[2]+' '+file.split('_')[3]+' '+file.split('_')[4])
    if not file.split('_')[0] in results:
        results[file.split('_')[0]] = {}
    results[file.split('_')[0]][file.split('_')[2]+' '+file.split('_')[3]+' '+file.split('_')[4]] = load_obj(os.path.join(base_path, file))
```

```python
print("{:<30} {:<23} {:<23}".format('', 'EER', 'FAR1%'))
print("{:<30} {:<11} {:<11} {:<11} {:<11}".format('', 'ANY', 'AVG', 'ANY', 'AVG'))
print("{:<30} {:<5} {:<5} {:<5} {:<5} {:<5} {:<5} {:<5} {:<5}".format('', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'))
out = []
for model, model_value in results.items():
    for mv, mv_value in model_value.items():
        eer_any_m = []
        eer_any_f = []
        eer_avg_m = []
        eer_avg_f = []
        far1_any_m = []
        far1_any_f = []
        far1_avg_m = []
        far1_avg_f = []
        for mv_file in mv_value.keys():
            mvtype = mv_file.split('_')[0]
            eer_any_m.append(len([1 for fac in mv_value[mv_file]['any']['eer']['m'] if fac > 0]) / 500)
            eer_any_f.append(len([1 for fac in mv_value[mv_file]['any']['eer']['f'] if fac > 0]) / 500)
            eer_avg_m.append(len([1 for fac in mv_value[mv_file]['avg']['eer']['m'] if fac > 0]) / 500)
            eer_avg_f.append(len([1 for fac in mv_value[mv_file]['avg']['eer']['f'] if fac > 0]) / 500)
            far1_any_m.append(len([1 for fac in mv_value[mv_file]['any']['far1']['m'] if fac > 0]) / 500)
            far1_any_f.append(len([1 for fac in mv_value[mv_file]['any']['far1']['f'] if fac > 0]) / 500)
            far1_avg_m.append(len([1 for fac in mv_value[mv_file]['avg']['far1']['m'] if fac > 0]) / 500)
            far1_avg_f.append(len([1 for fac in mv_value[mv_file]['avg']['far1']['f'] if fac > 0]) / 500)
        eer_any_m = float(np.mean(eer_any_m))
        eer_any_f = float(np.mean(eer_any_f))
        eer_avg_m = float(np.mean(eer_avg_m))
        eer_avg_f = float(np.mean(eer_avg_f))
        far1_any_m = float(np.mean(far1_any_m))
        far1_any_f = float(np.mean(far1_any_f))
        far1_avg_m = float(np.mean(far1_avg_m))
        far1_avg_f = float(np.mean(far1_avg_f))
        out.append([model, mv, eer_any_m, eer_any_f, eer_avg_m, eer_avg_f, far1_any_m, far1_any_f, far1_avg_m, far1_avg_f])
```

```python
df = pd.DataFrame(out, columns =['Model','MV Set',('EER','ANY','M'),('EER','ANY','F'),('EER','AVG','M'),('EER','AVG','F'), ('FAR1','ANY','M'),('FAR1','ANY','F'),('FAR1','AVG','M'),('FAR1','AVG','F')], dtype = float) 
df = df.sort_values(by=['Model','MV Set'])
df.columns = pd.MultiIndex.from_tuples([('-','-','Model'),('-','-','MV Set'),('EER','ANY','M'),('EER','ANY','F'),('EER','AVG','M'),('EER','AVG','F'), ('FAR1','ANY','M'),('FAR1','ANY','F'),('FAR1','AVG','M'),('FAR1','AVG','F')])
```
